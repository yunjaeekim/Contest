```python
OPST_target = pd.read_csv('OPST_target.csv', encoding='cp949')
```

```python
#1. LH데이터랑 크롤링 데이터 일치하는 명단(정답 Cost merge된것만)
OPST_target.drop(columns = ['cost'], inplace = True)
```

```python
OPST_target
```

```python
#2. 크롤링 데이터에서 LH 정답데이터에 있는 데이터 삭제
OPST_최종
```

```python
OPST_예측 = OPST_최종[~OPST_최종['opst'].isin(OPST_target['opst'])]
```

```python
OPST_예측
```

```python
OPST_예측['평수'] = OPST_예측['평수'].apply(lambda x : round(x))
OPST_예측
```

```python
OPST_cluster = OPST_예측[['floor', 'gu', 'dong', '평수']]
OPST_cluster
```

```python
from kmodes.kprototypes import KPrototypes
num_indices = [0,1,2]

kproto = KPrototypes(n_clusters= 20, init='Cao', verbose=1)
clusters = kproto.fit_predict(OPST_cluster, categorical=num_indices)

print(clusters)
```

```python
OPST_예측['Cluster'] = clusters
```

```python
OPST_예측
```
