```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.rc('font', family = 'malgun Gothic')
%matplotlib inline
import warnings
```

```python
speed = pd.read_excel('2023_차속도_1월.xlsx')
geocode_start = pd.read_excel('시점명.xlsx')
geocode_end = pd.read_excel('종점명.xlsx')
```

```python
speed.head()
```

```python
geocode_start
```

```python
geocode_start.drop("일자", axis = 1, inplace=True)
geocode_end.drop("일자", axis = 1, inplace=True)
```

```python
speed = speed.merge(geocode_start, on = '시점명');speed
```

```python
speed.rename(columns = {'Latitude':'Latitude_start','Longitude':'Longitude_start'}, inplace = True)
```

```python
speed = speed.merge(geocode_end,on = '종점명');speed
```

```python
speed.rename(columns = {'Latitude':'Latitude_end','Longitude':'Longitude_end'}, inplace = True)
```

```python
sp = speed[(speed['Latitude_start'].isnull() == False) | (speed['Latitude_end'].isnull() == False)];sp
```

```python
sp['speed_mean'] = sp[['01시', '02시', '03시', '04시', '05시', '06시', '07시', '08시', '09시', '10시', '11시', '12시', '13시', '14시', '15시', '16시', '17시', '18시', '19시', '20시', '21시', '22시', '23시', '24시']].mean(axis=1)
```

```python
sp = sp[['도로명','방향','거리','차선수','Latitude_start','Longitude_start','Latitude_end','Longitude_end','speed_mean']];sp
```

```python
sp.info()
```

```python
sp[['Latitude_start','Longitude_start','Latitude_end','Longitude_end']] = sp[['Latitude_start','Longitude_start','Latitude_end','Longitude_end']].fillna("0")
```

```python
sp
```

```python
sp[['Latitude_start','Longitude_start']] =  sp[['Latitude_start','Longitude_start']].replace({'Latitude': 0, 'Longitude' :0})
sp[['Latitude_end','Longitude_end']] =  sp[['Latitude_end','Longitude_end']].replace({'Latitude': 0, 'Longitude' :0})
```

```python
sp[sp['Latitude_end']=='Latitude']
```

```python
sp[['Latitude_start','Longitude_start']] = sp[['Latitude_start','Longitude_start']].astype(float)
sp[['Latitude_end','Longitude_end']] = sp[['Latitude_end','Longitude_end']].astype(float)
```

```python
sp.info()
```

```python
sp['Latitude_start1'] = sp.apply(lambda x: x['Latitude_start'] if x['Latitude_start'] != 0  else x['Latitude_end'], axis=1)
sp['Longitude_start1'] = sp.apply(lambda x: x['Longitude_start'] if x['Longitude_start'] != 0 else x['Longitude_end'], axis=1)

sp['Latitude_end1'] = sp.apply(lambda x: x['Latitude_end'] if x['Latitude_end'] != 0  else x['Latitude_start'], axis=1)
sp['Longitude_end1'] = sp.apply(lambda x: x['Longitude_end'] if x['Longitude_end'] != 0 else x['Longitude_start'], axis=1)

sp['Latitude']  = sp.apply(lambda x: (x['Latitude_start1'] + x['Latitude_end1']) / 2 , axis=1)
sp['Longitude'] = sp.apply(lambda x: (x['Longitude_start1'] + x['Longitude_end1']) / 2, axis=1)
```

```python
sp
```

```python
speed_final = sp.copy()
```

```python
speed_final
```

```python
speed_final = speed_final[['거리','차선수','Latitude','Longitude','speed_mean']];speed_final
```

```python
speed_final['speed_mean_차선수_mul'] = speed_final.apply(lambda x: x['차선수'] * x['speed_mean'] if x['차선수'] != 0 else x['speed_mean'], axis = 1)
```

```python
speed_final.to_csv("traffic_speed.csv", encoding = 'utf-8')
```
