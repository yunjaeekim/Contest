```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.rc('font', family = 'malgun Gothic')
%matplotlib inline
import warnings
```

```python
traffic = pd.read_excel('01월 서울시 교통량 조사자료(2023).xlsx', sheet_name=1)
speed = pd.read_excel('2023_차속도_1월.xlsx')
code = pd.read_excel('01월 서울시 교통량 조사자료(2023).xlsx', sheet_name=2);traffic.head(5)
```

```python
traffic['날짜'] = pd.to_datetime(traffic['일자'],format = '%Y%M%d')
traffic['month'] = traffic['날짜'].dt.month
traffic['year'] = traffic['날짜'].dt.year
```

```python
traffic.columns
```

```python
df = traffic.groupby(['지점명','방향'])[['0시', '1시', '2시', '3시', '4시',
       '5시', '6시', '7시', '8시', '9시', '10시', '11시', '12시', '13시', '14시', '15시',
       '16시', '17시', '18시', '19시', '20시', '21시', '22시', '23시']].sum().T.sum()
```

```python
traffic_cnt = pd.DataFrame(df, columns = ['교통량'])
```

```python
traffic_cnt['year'] = 2024
traffic_cnt['month'] = 1
traffic_cnt = traffic_cnt.reset_index();traffic_cnt
```

```python
code_cnt = code.copy()
code_cnt = code_cnt[['지점명칭','위도','경도']]
code_cnt.rename(columns = {'지점명칭':'지점명'}, inplace=True)
```

```python
traffic_cnt = traffic_cnt.merge(code_cnt, on = '지점명');traffic_cnt
```
