```python
people['교통수단타입명'] = people['교통수단타입명'].str[2:4];people
```

```python
move_people = people.merge(b_s, on = '역명'); move_people
```

```python
move_people.Latitude.value_counts().sort_index()
```

```python
move_people = move_people.query('교통수단타입명 == "간선"')
move_people.drop('역명', axis =1, inplace = True)
```

```python
a = move_people.groupby(['Latitude','Longitude'])[['총승객수_mean']].sum()
```

```python
a = a.query('37.413294 <= Latitude <= 37.715133 & 126.734086 <= Longitude <= 127.269311')
```

```python
# MinMAxScaler()
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
a['유동인구_mms'] = scaler.fit_transform(a['총승객수_mean'].values.reshape(-1, 1))
```

```python
final = a.reset_index()
```

```python
final.to_csv("유동인구(최종).csv", index = False, encoding = 'cp949')
```
