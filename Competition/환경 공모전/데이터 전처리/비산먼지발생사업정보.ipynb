```python
import pandas as pd
import numpy as np
```

```python
factory_22 = pd.read_csv("22년_비산먼지_발생사업장정보.csv", encoding = 'utf-8')
factory_23 = pd.read_csv("23년_비산먼지_발생사업장정보.csv", encoding = 'utf-8')
```

```python
factory_23
```

```python
factory_22['설치(공사)종료일'] = factory_22.apply(lambda x: x['데이터기준일자(신고일자)'] if x['설치(공사)종료일'] == '--' else x['설치(공사)종료일'],axis=1)
factory_22['설치(공사)시작일'] = factory_22.apply(lambda x: x['데이터기준일자(신고일자)'] if x['설치(공사)시작일'] == '--' else x['설치(공사)시작일'],axis=1)

factory_23['설치(공사)종료일'] = factory_23.apply(lambda x: x['데이터기준일자(신고일자)'] if x['설치(공사)종료일'] == '--' else x['설치(공사)종료일'],axis=1)
factory_23['설치(공사)시작일'] = factory_23.apply(lambda x: x['데이터기준일자(신고일자)'] if x['설치(공사)시작일'] == '--' else x['설치(공사)시작일'],axis=1)
```

```python
factory_22
```

```python
factory_22['종료_year'] = factory_22['설치(공사)종료일'].str[:4].astype(int)
factory_22['종료_month'] = factory_22['설치(공사)종료일'].str[5:7].astype(int)
factory_22['시작_year'] = factory_22['설치(공사)시작일'].str[:4].astype(int)
factory_22['시작_month'] = factory_22['설치(공사)시작일'].str[5:7].astype(int)

factory_23['종료_year'] = factory_23['설치(공사)종료일'].str[:4].astype(int)
factory_23['종료_month'] = factory_23['설치(공사)종료일'].str[5:7].astype(int)
factory_23['시작_year'] = factory_23['설치(공사)시작일'].str[:4].astype(int)
factory_23['시작_month'] = factory_23['설치(공사)시작일'].str[5:7].astype(int)

factory_22.head()
```

```python
factory_23 = factory_23.query('시작_year <= 2023 and 종료_year >= 2023')
factory_22 = factory_22.query('시작_year <= 2022 and 종료_year >= 2022')
```

```python
factory_23
```

```python
factory_23.fillna("000", inplace=True)
factory_22.fillna("000", inplace=True)
```

```python
factory_23 = factory_23.query('소재지도로명주소 != 000 or 소재지지번주소 != 000')
factory_22 = factory_22.query('소재지도로명주소 != 000 or 소재지지번주소 != 000')
```

```python
factory_22.to_csv("factory_22.csv", encoding = 'utf-8')
factory_23.to_csv("factory_23.csv", encoding = 'utf-8')
```
