```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
공사장_지수_standard = pd.read_csv("공사장_지수.csv", encoding = 'cp949')
교통량_지수_standard = pd.read_csv("교통량_지수.csv", encoding = 'cp949')
도로재비산먼지_지수 = pd.read_csv("도로재비산먼지_지수.csv", encoding ='cp949')
```

```python
도로재비산먼지_지수_standard.drop(columns= ['NO','측정일자'], axis =1)
```

```python
교통량_지수_standard
```

```python
scaler = StandardScaler()
공사장_지수_standard['공사장_지수_standard'] = scaler.fit_transform(공사장_지수_standard['공사장_지수'].values.reshape(-1, 1))
교통량_지수_standard['교통량_지수_standard'] = scaler.fit_transform(교통량_지수_standard['교통량'].values.reshape(-1, 1))
```

```python
공사장_지수_standard['공사장_지수_log'] = np.log(공사장_지수_standard['공사장_지수'])
교통량_지수_standard['교통량_지수_log'] = np.log(교통량_지수_standard['교통량'])
```

```python
공사장_지수_standard.drop('번호', axis =1, inplace = True)
```

```python
교통량_지수_standard.to_csv("교통량지수_standard.csv", index = False, encoding ='cp949')
공사장_지수_standard.to_csv("공사장_지수_standard.csv", index = False, encoding ='cp949')
```
