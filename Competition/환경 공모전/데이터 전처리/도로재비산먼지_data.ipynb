```python
import pandas as pd
import numpy as np
```

```python
기온_온도_23 = pd.read_csv("23_도로재비산먼지.csv", encoding = 'utf-8')
기온_온도_22 = pd.read_csv("22_도로재비산먼지.csv", encoding = 'utf-8')
```

```python
display(기온_온도_23)
display(기온_온도_22)
```

```python
print(기온_온도_23.info(),기온_온도_22.info())
```

```python
print(기온_온도_23.측정일자.nunique(),'\n',기온_온도_22.측정일자.nunique())
```

```python
weather = pd.concat([기온_온도_23,기온_온도_22], axis = 0).reset_index(drop = True)
```

```python
weather.head()
```

```python
weather['측정일자'].unique()
```

```python
weather['year'] = weather['측정일자'].str[0:4]
weather['month'] = weather['측정일자'].str[5:7]
```

```python
weather['측정시간'].unique()
```

```python
weather['측정시간'] = weather['측정시간'].str[:2]
```

```python
weather[weather['재비산먼지 평균농도(㎍/㎥)'] ==0].sum()
```

```python
weather['측정거리별_평균농도'] = weather.apply(lambda x: x['측정거리(km)']/ x['재비산먼지 평균농도(㎍/㎥)'] if x['재비산먼지 평균농도(㎍/㎥)'] != 0 else 0, axis=1)
weather.head()
```

### year, month, gu 별로 year이 기준이 되고 평균 농도와 오차 정도에 대한 칼럼 생성하기 위한 과정

```python
x = weather.groupby('year')['재비산먼지 평균농도(㎍/㎥)'].mean().reset_index(name='year_평균농도')
weather = weather.merge(x, on = ['year'])
x1 = weather.groupby(['year','month'])['재비산먼지 평균농도(㎍/㎥)'].mean().reset_index(name='year_month_평균농도')
weather = weather.merge(x1, on = ['year','month'])
x2= weather.groupby(['year','지역명'])['재비산먼지 평균농도(㎍/㎥)'].mean().reset_index(name='year_gu_평균농도')
weather = weather.merge(x2, on = ['year','지역명'])
weather.head()
```

```python
weather.head()
```

```python
weather['year_오차'] = weather.apply(lambda x: x['year_평균농도'] - x['재비산먼지 평균농도(㎍/㎥)'], axis = 1)
weather['year_month_오차'] = weather.apply(lambda x: x['year_month_평균농도'] - x['재비산먼지 평균농도(㎍/㎥)'], axis = 1)
weather['year_gu_오차'] = weather.apply(lambda x: x['year_gu_평균농도'] - x['재비산먼지 평균농도(㎍/㎥)'], axis = 1)
weather.head()
```

```python
weather_23 = weather[weather['year'] == '2023']
weather_22 = weather[weather['year'] == '2022']

weather_23
```

```python
weather_23.to_csv("도로재비산머지_22.csv", encoding = 'cp949', index = False)
weather_22.to_csv("도로재비산먼지_23.csv", encoding = 'cp949', index = False)
```
